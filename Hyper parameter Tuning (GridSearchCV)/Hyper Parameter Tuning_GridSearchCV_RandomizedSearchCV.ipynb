```python
import pandas as pd
df = pd.DataFrame(iris.data,columns = iris.feature_names)
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(iris.data,iris.target,test_size=0.3)
```

```python
from sklearn.svm import SVC
model = SVC(kernel='rbf',C=30,gamma='auto')
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

<h3>Applying K-Fold Cross Validation</h3>

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(SVC(kernel='linear',C=10,gamma='auto'),iris.data,iris.target,cv=5)
```

```python
cross_val_score(SVC(kernel='rbf',C=10,gamma='auto'),iris.data,iris.target,cv=5)
```

```python
cross_val_score(SVC(kernel='rbf',C=20,gamma='auto'),iris.data,iris.target,cv=5)
```

<h3>Applying K-Fold Cross Validation but by for loop</h3>

```python
import numpy as np
kernels = ['rbf','linear']
C = [1,10,20]
avg_score = []
for kval in kernels:
    for cval in C:
        cv_score = cross_val_score(SVC(kernel=kval,C=cval,gamma='auto'),iris.data,iris.target,cv=5)
        avg_score.append(np.average(cv_score))
avg_score
```

<h3>Grid Search CV gives exact same things like above cell</h3>

```python
from sklearn.model_selection import GridSearchCV
clf = GridSearchCV(SVC(gamma='auto'),{
    'C' :[1,10,20],
    'kernel' :['rbf','linear']

},cv=5, return_train_score=False)
```

```python
clf.fit(iris.data,iris.target)
```

```python
clf.cv_results_
```

```python
df = pd.DataFrame(clf.cv_results_)
df.head()
```

```python
df[['param_C','param_kernel','mean_test_score']]
```

```python
dir(clf)
```
